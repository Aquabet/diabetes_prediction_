```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import seaborn as sns
```

```python
data = pd.read_csv('diabetes.csv')
data.head()
```

```python
X = data[["Pregnancies", "Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI", "DiabetesPedigreeFunction", "Age"]]
y = data[["Outcome"]]
```

```python
pca = PCA(n_components=2)
pca.fit(X)
print (pca.explained_variance_ratio_) # Returns the proportion of each principal component of the PCA model
print (pca.explained_variance_) # How much variation in a dataset can be attributed to each of the principal components
```

```python
reduced_x = pca.transform(X)

y = np.array(y)
arr = np.concatenate((reduced_x, y), axis=1)

sns.scatterplot(data=arr, x=arr[:, 0], y=arr[:, 1], hue=arr[:, 2], marker='+')


# Above are new data
new_X = X[5:7]
new_y = y[5:7]

reduced_new_X = pca.transform(new_X)

new_arr = np.concatenate((reduced_new_X, new_y), axis=1)

sns.scatterplot(data=new_arr, x=new_arr[:, 0], y=new_arr[:, 1], hue=new_arr[:, 2], marker='o')
```
