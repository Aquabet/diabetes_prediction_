```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
data = pd.read_csv('diabetes.csv')
data.describe()
```

```python
Response_PredVars = data[['Pregnancies','Glucose','BloodPressure',
                          'SkinThickness','Insulin','BMI','DiabetesPedigreeFunction',
                          'Age','Outcome']]
Response_PredVars.corr()
```

```python
Response_PredVars = data[['Pregnancies','Glucose','BMI','DiabetesPedigreeFunction',
                          'Age','Outcome']]
Response_PredVars.corr()
```

```python
X = data[['Pregnancies','Glucose','BMI','DiabetesPedigreeFunction','Age',]]
Y = data['Outcome']
X,Y
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y,random_state=0,test_size=0.33)
print(x_train.shape, y_train.shape,x_test.shape,y_test.shape)
```

```python
from sklearn import linear_model
linreg = linear_model.LinearRegression()
linreg.fit(x_train, y_train)
```

```python
import statsmodels.formula.api as smf
trainDataExcel = pd.concat([x_train, y_train], axis=1)
testDataExcel = pd.concat([x_test, y_test], axis=1)
mlr = smf.ols(formula='Outcome ~ Pregnancies + Glucose + BMI + DiabetesPedigreeFunction + Age', data=trainDataExcel).fit()
print(mlr.summary())
```

```python
mlr = smf.ols(formula='Outcome ~ Pregnancies + Glucose + BMI + DiabetesPedigreeFunction + Age', data=testDataExcel).fit()
print(mlr.summary())
```
